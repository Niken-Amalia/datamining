# Proses diskritisasi dengan equal width dan equal frequency

### Diskritisasi
Diskritisasi digunakan untuk mengubah atribut numerik menjadi atribut kategorikal. Perubahan tersebut dilakukan dengan mengkategorikan atribut numerik menjadi beberapa tingkatan atribut kategorikal. Dalam implementasinya, diskritisasi memiliki dua pendekatan, equal-width intervals dan equal-frequency intervals.

### Equal-width intervals adalah diskritisasi yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar. Fungsi cut digunakan untuk mencari interval menggunakan metode Equal Width Intervals

#### Sepal Width
Equal Width Intervals lebar kelopak bunga Iris

Dikelompokkan menjadi 3 kategori yaitu Kecil, Sedang, dan Lebar

```python
# label kategori Sepal Width
labelK = ["Kecil", "Sedang", "Lebar"]
# Untuk menghitung banyaknya kategori label Sepal Width
jumlahK = len(labelK)

# untuk menghitung interval equal-width
dt_lebarsepal_ew = pd.cut(lebar_sepal, jumlahK, True, labelK)
jumlahdt_kategori_lebarsepal_ew = dt_lebarsepal_ew.value_counts()
jumlahdt_interval_lebarsepal_ew = pd.cut(lebar_sepal, jumlahK, True).value_counts()
```

```python
# data dari Sepal Width beserta kategorinya
df_lebarsepal_ew=pd.DataFrame(pd.concat((lebar_sepal, dt_lebarsepal_ew), axis = 1))
df_lebarsepal_ew.columns = ["SepalWidth", "category"]
df_lebarsepal_ew
```

## Entropi
Entropi adalah nilai informasi yang menyatakan
ukuran ketidakpastian(impurity) dari atribut dari
suatu kumpulan obyek data dalam satuan bit.

$$
\begin{align*}
\displaystyle Entropy(S) &= \sum_{i=0}^{n} -phi \ log_{2} \ phi \\
Keterangan &: \\
S &= Himpunan \ dataset  \\
n &= Banyaknya \ partisi \ S \\
phi &= Proporsi \ dari \ S_{i} \ terhadap \ S
\end{align*}
$$

```python
# Data yang akan digunakan
targetdata = df_lebarsepal_ew
# Label Kategori
label = ['Kecil', 'Sedang', 'Lebar']
```

```python
#fungsi yang digunakan untuk menghitung banyak data berdasarkan kategori
def hitung_kategori(kolom_data, label, kolom, kategori):
  grup = kolom_data.groupby(kategori).count()
  jumlah_kategori = []
  for label in label:
    jumlah_kategori.append(grup.loc[label, kolom])
  return jumlah_kategori
  
```

```python
hitung_kategori(targetdata, label, targetdata.columns[0], targetdata.columns[1])
```

```python
#fungsi menghitung nilai entropy
def entropy(targetdata):
    all_prob = []
    for prob in targetdata:
        if (prob/sum(targetdata) != 0):
            all_prob.append(prob/sum(targetdata) * log2(prob/sum(targetdata)))
        else:
            all_prob.append(0)
    return -(sum(all_prob))
```

```python
#fungsi yang digunakan untuk proses info D
def split(nilai,kolom_data, label, kolom, kategori):
  less_group =kolom_data[kolom_data[kolom] < nilai]
  greater_group =kolom_data[kolom_data[kolom] >= nilai]

  length_less_group = hitung_kategori(less_group, label, kolom, kategori)
  length_greater_group = hitung_kategori(greater_group, label, kolom, kategori)

  return (length_less_group, length_greater_group)
```

### Info

$$ Info_A(D) = \frac{|D_1|}{|D|} Entropy (D_1)\frac{|D_2|}{|D|}+ Entropy (D_2)$$

```python
# fungsi menghitung nilai info
def info(d, targetdata):
    temp = []
    for value in d:
        temp.append((sum(value) / targetdata.shape[0]) * entropy(value))
    return sum(temp)
```

```python
# fungsi menghitung nilai gain
def gain(inisial, new):
  return inisial - new
```
